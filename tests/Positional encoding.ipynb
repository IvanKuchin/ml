```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
position = 5
d_model = 8

position_array = np.arange(position)[:, np.newaxis]
dim_array = np.arange(d_model)[np.newaxis, :]
```

```python
print("position_array:", position_array)
print("dim_array:", dim_array)
```

```python
def get_angle_rates(positions, dimensions):
    pos = np.arange(positions)[:, np.newaxis]
    dim = np.arange(dimensions)[np.newaxis, :]
    return pos / np.power(10000, (2. * (dim // 2) / np.float32(dimensions)))
```

```python
print(get_angle_rates(5, 10))
```

```python
plt.imshow(get_angle_rates(20,32))
```

```python
def positional_encoding(position, d_model):
    angle_rads = get_angle_rates(position, d_model)
    
    angle_rads[:, 0::2] = np.sin(angle_rads[:, 0::2])
    angle_rads[:, 1::2] = np.sin(angle_rads[:, 1::2])

    return angle_rads
```

```python
plt.imshow(np.rot90(positional_encoding(256, 128)))
```
